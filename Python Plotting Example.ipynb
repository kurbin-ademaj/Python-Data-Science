```python
bins = [30,40,50,60,70,80,90,100]
plt.hist(data_fifa.Overall, bins=bins, label='Skill Level')
plt.title('Distribution of Player Skills in FIFA')
plt.xlabel('Skill level')
plt.ylabel('Number of players')
plt.xticks(bins)
plt.legend()
plt.show()
```

```python
data_fifa[['Overall', 'Potential']].plot.hist(bins=20, alpha=0.6, figsize=(10,6))
plt.title('Overall vs Potential Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Players')
plt.show()
```

```python
# Count the number of players for each preferred foot
foot_counts = data_fifa['Preferred Foot'].value_counts()

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    foot_counts, 
    labels=foot_counts.index, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['skyblue', 'lightgreen'], 
    shadow=True
)

plt.title('Preferred Foot Distribution')
plt.show()
```

```python
# Safely clean and convert 'Weight' column
data_fifa['Weight'] = (
    data_fifa['Weight']
    .astype(str)            # ensure all values are strings
    .str.replace('lbs', '') # remove the 'lbs' text
    .str.strip()            # remove any spaces
)

# Convert cleaned weights to numbers (coerce invalid values to NaN)
data_fifa['Weight'] = pd.to_numeric(data_fifa['Weight'], errors='coerce')

# Define bins (weight ranges) and labels
bins = [0, 140, 160, 180, 200, 250]
labels = ['<140 lbs', '140-160 lbs', '160-180 lbs', '180-200 lbs', '>200 lbs']

# Create weight categories
data_fifa['Weight Category'] = pd.cut(data_fifa['Weight'], bins=bins, labels=labels, right=False)

# Count players in each category
weight_counts = data_fifa['Weight Category'].value_counts().sort_index()

# Add 'explode' to highlight a slice
explode = [0, 0, 0.1, 0, 0]  # highlight 160180 lbs category

# Plot the pie chart
plt.figure(figsize=(7, 7))
plt.pie(
    weight_counts,
    labels=weight_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.Paired.colors,
    shadow=True,
    explode=explode
)
plt.title('Player Weight Distribution')
plt.show()
```

```python
barcelona = data_fifa.loc[data_fifa.Club == 'FC Barcelona']['Overall']
madrid = data_fifa.loc[data_fifa.Club == 'Real Madrid']['Overall']
```
